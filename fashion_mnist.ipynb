# CASE STUDY: FASHION CLASS CLASSIFICATION

We wanted to build kind of an app or a model that can look at images and can tell us exactly what category in this image. Is it like a short is it a bag? Is it like a hat and so on.
So we wanted to build kind of artificial intelligence based deep learning model that can classify images into different categories or different classes.

# STEP #1: PROBLEM STATEMENT AND BUSINESS CASE

Fashion training set consists of 70,000 images divided into 60,000 training and 10,000 testing samples. Dataset sample consists of 28x28 grayscale image, associated with a label from 10 classes.

The 10 classes are as follows:  
0 => T-shirt/top
1 => Trouser
2 => Pullover
3 => Dress
4 => Coat
5 => Sandal
6 => Shirt
7 => Sneaker
8 => Bag
9 => Ankle boot

Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255.

# STEP #4: TRAINING THE MODEL

```python
# Prepare the training and testing dataset
X_train = training[:,1:]/255            #normalization used   . The pixel values are normalized to the range [0, 1] to improve the performance of the machine learning mode
y_train = training[:,0]

X_test = testing[:,1:]/255
y_test = testing[:,0]
```

```python
from sklearn.model_selection import train_test_split
# validation dataset that might used to help the model to generalize
X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.2, random_state = 12345)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# * unpack the tuple. Reshape the data in a form that CNN can accept
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))
```

```python
import keras # open source Neural network library 

# y_train = keras.utils.to_categorical(y_train, 10)
# y_test = keras.utils.to_categorical(y_test, 10)
```

```python
# Import train_test_split from scikit library
# Import Keras
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam  #used to update the weights of the neural network during training, based on the gradients of the loss function with respect to the weights. 
from keras.callbacks import TensorBoard
```

```python
# build your CNN model using keras

# 1. Initialize your model

cnnModel = Sequential()

# 2. Start with Convoltional layer *

cnnModel.add(Conv2D(filters=64, kernel_size=(3,3), input_shape = (28,28,1), activation = 'relu'))

# Apply the non linearity
# 3. pooling
cnnModel.add(MaxPooling2D(pool_size=(2,2)))

# 4. faltening
cnnModel.add(Dropout(0.2))
cnnModel.add(Flatten())

# 5. Dense layer (think about number layers & what are the activation fucntions)

cnnModel.add(Dense(units=64, activation = 'relu'))
cnnModel.add(Dense(units=32, activation = 'relu'))
cnnModel.add(Dense(units=10, activation = 'sigmoid'))

#5. Complie the model by defineing the cost function and the optimizer for the model as well the used metrics to measure the qulaity of the model

cnnModel.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 6. Define the training parameters
    # number of epochs
    # batch size
    # traning and validation set

cnnModel.fit(x = X_train,
             y = y_train,
             batch_size= 512,
             epochs= 50,
             verbose=1,
             validation_data=(X_validate, y_validate)
            )
```

```python
# use sparse_categorical_crossentropy for the loss function and accuracy for the metrics of accuracy
```

# STEP #5: EVALUATING THE MODEL

```python
# 1. evaluate your model using test_set

accuracy = cnnModel.evaluate(X_test, y_test)
print('Test Acuuracy : {}'.format(accuracy[1]))


# Get the prediction classes for test data
predict_x = cnnModel.predict(X_test)
classes_x = np.argmax(predict_x,axis=1)
```

```python
L = 5
W = 5
fig, axes = plt.subplots(L, W, figsize = (12,12))
axes = axes.ravel() #

for i in np.arange(0, L * W):
    axes[i].imshow(X_test[i].reshape(28,28))
    axes[i].set_title("Prediction Class = {:0.1f}\n True Class = {:0.1f}".format(classes_x[i], y_test[i]))
    axes[i].axis('off')

plt.subplots_adjust(wspace=0.5)
```

```python
from sklearn.metrics import confusion_matrix
# use seaborn library to plot the confusion matrix

con_mat = confusion_matrix(y_test, classes_x)

# normalization
con_mat_norm = con_mat.astype('float') / con_mat.sum(axis=1)[:, np.newaxis]
con_mat_norm = np.around(con_mat_norm, decimals=2)

# === plot ===
plt.figure(figsize=(10, 10))
sns.heatmap(con_mat_norm, annot=True, cmap='Blues')

plt.ylim(0, 10)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()

# Sum the diagonal element to get the total true correct values
```

```python
from sklearn.metrics import classification_report

num_classes = 10
target_names = ["Class {}".format(i) for i in range(num_classes)]

print(classification_report(y_test, classes_x, target_names = target_names))
```
